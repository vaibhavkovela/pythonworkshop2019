## use of lambda with filter

```python
 filterlist(li):
    a=[]
    for i in li:
        if i%2==0:
            a.append(i)
    return a
li=[1,2,3,4,5,6,7]
filterlist(li)
```

```python
#lambda as filter
li=[1,2,3,4,5,67,8]
lali=list(filter(lambda x: (x%2==0),li))
print(lali)
```

### lambda with map()
    -map() functin in python takes in a function and list as argument

```python
def squarelist(li):
    a=[]
    for i in li:
        a.append(i**2)
    return a
li={1,2,3,4,5,6}
squarelist(li)
```

```python
#lambda with map
li=[1,2,3,4,5,6,7,8]
mali=list(map(lambda x:x**2,li))
print(mali)
```

## lambda with reduce()
 

```python
def sumlist(li):
    s=0
    for i in li:
        s+=i
    return s
li=[1,2,3,4,5,6]
sumlist(li)
```

```python
from functools import reduce
li=[1,2,3,4,5,6,7,8,9]
s=reduce((lambda x,y:x+y),li)
print(s)
```
