## visualizing data
    -an interesting way to display dataframe
    -with matploitb

```python
from matplotlib.pyplot import *
a1=[1,20,3,40]
a2=[5,60,7,80]
figure(figsize=(4,4))
plot(a1,a2,color='red')
xlabel('x-axis')
ylabel('y-axis')
show()
```

```python
marks=[99,99,100,99]
avd_marks=[90,85,80,81]
divisions=[1,2,3,4]
division=[1.5,2.5,3.5,4.5]
bar(divisions,marks,color='blue',width=0.50)
bar(division,avd_marks,color='blue',width=0.40)
```

```python

barh(divisions,marks,color='blue')
barh(division,avd_marks,color='red')
```

```python
import matplotlib.pyplot as plt
lang=['python','java','c','c++']
rating=[65,70,60,75]
pie(rating,labels=lang,shadow=True,startangle=90,autopct='%1.1f%%')
plt.legend(loc='best')
show()
```
